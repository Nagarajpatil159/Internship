```python
# accessing search_bar and searching
search_bar=driver.find_element(By.XPATH,"/html/body/div[2]/div[2]/form/div[1]/div[1]/div[2]/div/div[2]/textarea")
search_bar.send_keys("cars")
```

```python
# clicking on search button

search_button=driver.find_element(By.XPATH,"/html/body/div[2]/div[2]/form/div[1]/div[1]/div[2]/button/div/span")
search_button.click()
```

```python
cars=[]
```

```python
try:
    car=driver.find_elements(By.XPATH,'//div[@class="H8Rx8c"]//img')
    for i in car:
        cars.append(i.get_attribute('src'))
except NoSuchElementException as e:
        cars.append("-")
```
