```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
# opening chrome

driver=webdriver.Chrome()
```

```python
ranks=[]
names=[]
artists=[]
dates=[]
views=[]
```

```python
# scraping names

for page in range(0,1):
    try:
        name=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]//tbody/tr/td[1]')
        
        for i in name:
            names.append(i.text)  # appending names
            
    except NoSuchElementException as e:    # handling exception
        names.append("-")
    
```

```python
# scraping views

for page in range(0,1):
    try:
        view=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]//tbody/tr/td[3]')
        
        for i in view:
            views.append(i.text)  # appending views
            
    except NoSuchElementException as e:    # handling exception
        views.append("-")
    
```

```python
# scraping dates

for page in range(0,1):
    try:
        date=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]//tbody/tr/td[4]')
        
        for i in date:
            dates.append(i.text)  # appending dates
            
    except NoSuchElementException as e:    # handling exception
        dates.append("-")
    
```

```python
df=pd.DataFrame({'Names':names,'published_by':artists,'published_on':dates,'Views_in_billions':views})
df
```
